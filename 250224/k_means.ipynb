```python
# Gerekli ktphaneleri ykleyelim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1. Adm: Iris veri setini ykle
iris = load_iris()
X = iris.data  # zellikler
y = iris.target  # Gerek etiketler (k-means iin gerekmez)

# Veriyi standartlatr
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Adm: K-means algoritmasn uygula
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 kmeye ayracaz
kmeans.fit(X_scaled)

# 3. Adm: Sonular grselletir
plt.figure(figsize=(8, 6))

# Kmeleri grselletirelim
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis', marker='o')

# Kme merkezlerini iaretleyelim
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='red', marker='X', label='Merkezler')

plt.xlabel('zellik 1')
plt.ylabel('zellik 2')
plt.title('K-means Kmelenmesi (2 zellik Grselletirme)')
plt.legend()
plt.show()

# Kmelerin hangi etiketlere ait olduunu grelim
print(f"Kmelenmi Etiketler: {kmeans.labels_}")
```
