```python
# Gerekli kütüphaneleri yükleyelim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1. Adım: Iris veri setini yükle
iris = load_iris()
X = iris.data  # Özellikler
y = iris.target  # Gerçek etiketler (k-means için gerekmez)

# Veriyi standartlaştır
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Adım: K-means algoritmasını uygula
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 kümeye ayıracağız
kmeans.fit(X_scaled)

# 3. Adım: Sonuçları görselleştir
plt.figure(figsize=(8, 6))

# Kümeleri görselleştirelim
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis', marker='o')

# Küme merkezlerini işaretleyelim
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='red', marker='X', label='Merkezler')

plt.xlabel('Özellik 1')
plt.ylabel('Özellik 2')
plt.title('K-means Kümelenmesi (2 Özellik Görselleştirme)')
plt.legend()
plt.show()

# Kümelerin hangi etiketlere ait olduğunu görelim
print(f"Kümelenmiş Etiketler: {kmeans.labels_}")
```
